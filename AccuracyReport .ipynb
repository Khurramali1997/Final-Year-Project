```python
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
path = "/content/drive/MyDrive/FYP/"
```

```python
pred = pd.read_csv(f'{path}predictions.csv')
test_labels = pd.read_csv(f'{path}test_labels.csv')
```

```python
test_labels = test_labels[test_labels['toxic'] != -1]
```

```python
pred
```

```python
test_labels
```

```python
tl = test_labels['id'].tolist()
```

```python
pred = pred[pred['id'].isin(tl)]
```

```python
# test_labels.to_csv('test_label_com.csv',index=False)
# pred.to_csv('pred_com.csv',index=False)
```

```python
attr = 'obscene'
print("Confusion Matrix : \n")
print(confusion_matrix(pred[attr],test_labels[attr]))
print("\nClassification Report :\n")
print(classification_report(pred[attr],test_labels[attr]))
```

```python
print("Accuracy based on indiviual attribute :\n")
tox_acc = accuracy_score(pred['toxic'],test_labels['toxic'])
print("Toxic - " + str(tox_acc))

sev_acc = accuracy_score(pred['severe_toxic'],test_labels['severe_toxic'])
print("Severe_Toxic -" + str(sev_acc))

obs_acc = accuracy_score(pred['obscene'],test_labels['obscene'])
print("Obscene - " + str(obs_acc))

thr_acc = accuracy_score(pred['threat'],test_labels['threat'])
print("Threat - " + str(thr_acc))

ins_acc = accuracy_score(pred['insult'],test_labels['insult'])
print("Insult - " + str(ins_acc))

idh_acc = accuracy_score(pred['identity_hate'],test_labels['identity_hate'])
print("Identity_Hate - " + str(idh_acc))
```

```python
avg_acc = (tox_acc+sev_acc+obs_acc+thr_acc+ins_acc+idh_acc)/6.0
print("Average Accuracy - " + str(avg_acc))
```
